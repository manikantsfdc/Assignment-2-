```python
#import libraries 
from bs4 import BeautifulSoup
import requests

response = requests.get('https://www.amazon.in/s?k=laptop&i=computers&rh=n%3A1375424031%2Cp_n_feature_thirteen_browse-bin%3A12598163031%7C16757432031&dc&qid=1604810434&rnid=12598141031&ref=sr_nr_p_n_feature_thirteen_browse-bin_16 ')
product_html = BeautifulSoup(response.content, 'html.parser')

print(response.status_code)

headers = product_html.find_all('div', attrs = {'class': 'a-section a-spacing-medium'})
# print(headers)
for box in headers:
    # print("Title: " + box.find('b').text)
    # print("Description: " + box.find_all('div')[-1].text)
    title = box.find('h2')
    print(title.text)
    price = box.find('span', attrs = {'class': 'a-price-whole'})
    print(price.text)
    star = box.find_all('i')[0]
    print(star.text)

```
